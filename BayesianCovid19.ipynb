# 8. Data Cleaning and Visualization

```python
import numpy as np
import pymc3 as pm
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
import requests
import io
    
# Downloading the csv file from your GitHub account

url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))

# Printing out the first 5 rows of the dataframe
df.head()
```

```python

cal_cumsum = df[df["state"] == "California"].sort_values(by="date")
# Filter out data with less than 100 cases
cal_cumsum = cal_cumsum[cal_cumsum["cases"] >= 100]
cal_cumsum = cal_cumsum.drop(columns = ['fips'])
cal_cumsum.plot(logy=True)
plt.gca().set(xlabel="Date", ylabel="Count", title="California")
plt.show()


```

# Bayesian Exponential Infection Model

Exponential growth rates
There are a variety of fitted growth rates appearing in the literature. Could spend all day reading about it. It looks like a value of 0.2 is appropriate for larger countries.

India (0.18-0.19), Ganesh Kumar et al., http://arxiv.org/abs/2003.12017
China (0.24), Wu et al., https://arxiv.org/pdf/2003.05681.pdf
China (0.22), Batista, https://www.medrxiv.org/content/10.1101/2020.03.11.20024901v2.full.pdf

```python
state = "california"
days_since_100 = range(len(cal_cumsum))


# Create PyMC3 context manager
with pm.Model() as model:
  t = pm.Data(state+"x_data", days_since_100)
  confirmed_cases = pm.Data(state + "y_data", cal_cumsum["cases"].astype('float64').values)

  # Intercept - We fixed this at 100 since we cut off at 100. 
  a = pm.Normal("a", mu=100, sigma=10)

  # Slope - Growth rate: 0.2 seems to be the normal 
  b = pm.Normal("b", mu=0.2, sigma=0.5)

  # Exponential regression
  growth = a * (1 + b) ** t

  # Likelihood error
  eps = pm.HalfNormal("eps")

  # Likelihood - Counts here, so poission or negative binomial. Causes issues. Lognormal tends to work better?
  pm.Lognormal(state, mu=np.log(growth), sigma=eps, observed=confirmed_cases)

  trace = pm.sample()
  post_Pred = pm.sample_posterior_predictive(trace)
```

```python
pm.traceplot(trace)
plt.show()
```
